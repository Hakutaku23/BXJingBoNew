```python
map_list = {
    "卤化工段胶液总量2": [30000, 65000],
    "反应溴添加量": [200,500],
    "储罐胶浓在线检测": [13,25],
    "R510A温度": [38, 46],
    "R511A温度": [19,25],
    "R512A温度": [39,54],
    "R510B温度": [15, 40],
    "R511B温度": [35, 45],
    "R512B温度": [40, 55],
    "硬脂酸钙加注量": [700, 1300],
    "ESBO加注量": [100, 170],
    "中和碱液添加量": [2700, 7000],
    "R513温度": [40, 55],
    "R514温度": [40, 55]
}

def get_map_safe(df: pd.DataFrame, col_name: str, map_list: dict) -> pd.DataFrame:
    """
    清洗界外值：强校验数据类型，并将低于下限与高于上限的值记为np.nan
    """
    # 1. 强制转换为浮点数，将无法解析的异常字符串自动转为 NaN（防 Object 陷阱）
    df[col_name] = pd.to_numeric(df[col_name], errors='coerce')
    
    # 2. 获取阈值
    lower, upper = map_list[col_name]
    
    # 3. 边界过滤
    df.loc[df[col_name] < lower, col_name] = np.nan
    df.loc[df[col_name] > upper, col_name] = np.nan

    return df
```

```python
df = df.set_index("time")
df
```

```python
def apply_global_map(df: pd.DataFrame, map_dict: dict) -> pd.DataFrame:
    """
    遍历 DataFrame 所有列，将超出 map_dict 边界的值置为 NaN
    """
    # 建立副本，彻底阻断 SettingWithCopyWarning 及底层数据污染
    df_clean = df.copy()
    
    # 仅遍历存在于规则字典中的列，实现解耦
    for col in df_clean.columns:
        if col in map_dict:
            # 1. 强转为数值型，规避隐式 Object 字符串比较陷阱
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
            
            # 2. 提取上下限
            lower, upper = map_dict[col]
            
            # 3. 向量化掩码操作：超出边界的布尔矩阵为 True，mask 会自动将其替换为 NaN
            outlier_condition = (df_clean[col] < lower) | (df_clean[col] > upper)
            df_clean[col] = df_clean[col].mask(outlier_condition)
            
    return df_clean
df_cleaned = apply_global_map(df, map_list)

# 查看清洗后的结果
display(df_cleaned.head())
```
